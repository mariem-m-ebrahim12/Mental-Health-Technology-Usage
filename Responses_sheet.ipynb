# **#solve the columns problems**

```python
df = df.drop(['Unnamed: 15', 'Unnamed: 16','Unnamed: 17','@'], axis=1)
df
```

```python
df.shape                  # all the empity colums dropped
```

 Fill 'Technology_Usage_Hours','Social_Media_Usage_Hours','Gaming_Hours', 'Sleep_Hours','Physical_Activity_Hours' with numeric value

```python
# Fill the column with numeric value
df['Technology_Usage_Hours'].replace({'2 - 4 Hours': (2+3+4)/3, '4 - 6 Hours': (4+5+6)/3, '6 - 8 Hours': (6+7+8)/3, '8 - 10 Hours' : (8+9+10)/3}, inplace=True)
df['Social_Media_Usage_Hours'].replace({'2 - 4 Hours': (2+3+4)/3, '4 - 6 Hours': (4+5+6)/3 }, inplace=True)
df['Gaming_Hours'].replace({'1 - 2 Hours': (0+1+2)/3, '2 - 4 Hours': (2+3+4)/3,'4 - 6 Hours' : (4+5+6)/3,'6 - 8 Hours' : (6+7+8)/3}, inplace=True)
df['Physical_Activity_Hours'].replace({'2 -3 Hours': (0+1+2+3)/4, '3 - 4 Hours': (3+4)/2,'5 - 6 Hours' : (5+6)/2,'2 -3  Hours': (2+3/2)}, inplace=True)
df['Sleep_Hours'].replace({'4 - 6 Hours' : (4+5+6)/3,'6 - 8 Hours' : (6+7+8)/3,'8 - 10 Hours': (8+9+10)/3}, inplace=True)
df.tail()
```

```python
df['Screen_Time_Hours'].unique()               #There are some strings must be replaced with numeric values
```

```python
#strings replaced by numeric values
df['Screen_Time_Hours'].replace({'1-2 Hours': (1+2)/2, '4-6 Hours ': (4+5+6)/3,'6-8 Hours': (6+7+8)/3,'6_8 hours': (6+7+8)/3 }, inplace=True)
df['Screen_Time_Hours'].replace({'6 Hours' : 6,'3 hour' : 3,'7hr ': 7, '7hr':7,'2 hours': 2, '4 hours' : 4,'8 hours':8, '10 hours':10},inplace=True )
df['Screen_Time_Hours'].unique()                   #all values become numeric
```

```python
df['Age'].unique()                               # '22!' and '21years ' must be replaced with numeric value
```

```python
df['Age'].replace({'21years ': 21, '22!': 22} , inplace=True)
df['Age'].unique()                               #all values become numeric
```

```python
df.columns                                     # know the columns name to change its data type
```

```python
#change data type
df['Age'] = df['Age'].astype(int)
df['Gaming_Hours'] = df['Gaming_Hours'].astype(float)
df['Sleep_Hours'] = df['Sleep_Hours'].astype(float)
df['Physical_Activity_Hours'] = df['Physical_Activity_Hours'].astype(float)
df['Screen_Time_Hours'] = df['Screen_Time_Hours'].astype(float)
df['Technology_Usage_Hours'] = df['Technology_Usage_Hours'].astype(float)
df['Social_Media_Usage_Hours'] = df['Social_Media_Usage_Hours'].astype(float)
df.dtypes
```

```python
df['Gender'].unique()
```

```python
# Replace specific values in the 'Gender' column
df['Gender'] = df['Gender'].replace({
    'ÿßŸÜÿ´Ÿâ': 'Female',
    'F': 'Female',
    'Egyptian': 'Female',
    'Female ': 'Female',  # To handle any extra spaces
    'Male ': 'Male',      # To handle any extra spaces
    'Hghj': 'Male', # Example to replace invalid entries
    'ÿ∞ŸÉÿ±': 'Male',         # Arabic for Male

})

# Ensure only 'Male' or 'Female' are in the Gender column
df['Gender'] = df['Gender'].apply(lambda x: 'Male' if x in ['Male', 'ÿ∞ŸÉÿ±'] else 'Female' if x in ['Female', 'ÿßŸÜÿ´Ÿâ'] else 'Other')
df['Gender'].unique()
```

# **#Solve NULL Problem**

```python
# Use .any() to check if any value in the column is null
if df['Screen_Time_Hours'].isnull().any():
    # Calculate the sum of the three columns
     df['Screen_Time_Hours'] = df['Screen_Time_Hours'].fillna(df['Technology_Usage_Hours'] + df['Social_Media_Usage_Hours'] + df['Gaming_Hours'])
df['Screen_Time_Hours']
```

```python
#mode of the Support_Systems_Access
df['Online_Support_Usage'].mode()[0]
```

```python
# Fill empty cells in the 'Online Support Usage' column with 'Yes'
df['Online_Support_Usage'].fillna('Yes', inplace=True)
```

```python
#mode of the Support_Systems_Access
mode_Support_Systems_Access = df['Support_Systems_Access'].mode()[0]
mode_Support_Systems_Access
```

```python
#Fill null value of the Support_Systems_Access with the mode
df['Support_Systems_Access'] = df['Support_Systems_Access'].fillna(mode_Support_Systems_Access)
```
